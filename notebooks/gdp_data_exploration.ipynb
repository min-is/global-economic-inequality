```python
# import modules
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/min-is/global-economic-inequality/refs/heads/main/data/raw/gdpcapita.csv'
df = pd.read_csv(url, skiprows=4)
df.drop(df.columns[len(df.columns)-1], axis=1, inplace=True)
df.head(2)
```

```python
year_cols = [str(y) for y in range (1960, 2025)]
df.columns = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'] + year_cols
df.head()
```

```python
id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
df = df.melt(id_vars = id_vars, var_name='Year', value_name='GDP_per_capita')
df['Year'] = df['Year'].astype(int)
df['GDP_per_capita'] = pd.to_numeric(df['GDP_per_capita'], errors='coerce')
df.head()
```

```python
import perfplot

def setup(n):
    df = pd.DataFrame(np.random.randn(n))
    df[df > 0.9] = np.nan 
    return df

def isnull_any(df):
    return df.isnull().any()

def isnull_values_sum(df):
    return df.isnull().values.sum() > 0

def isnull_sum(df):
    return df.isnull().sum() > 0

def isnull_values_any(df):
    return df.isnull().values.any()

n_range = [2 ** k for k in range(15, 20)]  

perfplot.save(
    "out_gdp.png",
    setup=setup,
    kernels=[isnull_any, isnull_values_sum, isnull_sum, isnull_values_any],
    n_range=n_range,  
)
```
